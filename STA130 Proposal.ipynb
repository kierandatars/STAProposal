Research Question 1: 
Does receiving hugs have a measurable effect on increasing happiness levels in individuals?  
Variables  
1. The variables you plan to explore - 
Hug frequency would be the independent variable, which would be categorical on how 
often the individual receives hugs. I would split it up into two groups, my control group 
that did not receive hugs and the experimental group that received hugs. Happiness 
level would be the dependent variable being a numeric variable measured from 1 to 5.  
2. The analysis you plan to perform - 
Analysis 
I would create a two-sample bootstrapped confidence interval estimating the difference 
in average happiness change between the two groups. Involving: resampling with 
replacement from each group’s happiness change score. I would calculate the mean 
difference between the resampled groups. Then construct a confidence interval based 
on the distribution of these mean differences.  
3. the possible results your analysis may produce - 
The possible results could be a statistically significant difference in which the confidence 
interval for the difference in happiness change does NOT include zero, suggesting that there 
a sign. If the confidence interval does include zero, it indicates no statistically significant 
difference, suggesting hugs may not have a measurable impact on happiness.  
4. how this could be relevant to the objectives of the course project - 
The course project goal is to raise awareness about mental health, so if hugs make a person 
happy it should be important to raise awareness about spreading that.   
Hypothesis and how the result would be relevant  
5. Hypothesis  - 
Individuals who receive hugs will experience a greater increase in self-reported happiness 
compared to those who do not receive hugs. 
6. How will the result be relevant - 
If receiving hugs increases happiness, I believe that is important information that should 
help be encouraged as a mental health resource.
